```python
import matplotlib.pyplot as plt

def analyze_airports(airport): # This function takes the following functions and combines them so only one function will need to me run
    late_flights, ontime_flights, nodata_flights = flight_delay('C:\\Users\major\OneDrive\Desktop\\flights.May2017-Apr2018.csv', airport)
    graph_flights(late_flights, ontime_flights, nodata_flights)

def flight_delay(file, airport): #This defines our first function where we are trying to give the computer a file and the airport we are looking for

    try:  #Here we try to open the file given to the machine 
        fhand = open(file, 'r')
    except ImportError:
        exit()

#The following code basically takes the file given to it and then looks down two columns to look at the flights                                  delayed by weather and their associated airport and prints out all the late, on time and, no data flights for t                                  that particular airport 

    late_flights=0
    ontime_flights=0
    nodata_flights=0
    for line in fhand:          
        delimiter = ',' 
        t = line.split(delimiter) 
        if t[2].strip('"') in airport :
            try :
                if float (t[23]) > 0.00 :
                    #print('Late', airport)
                    late_flights= late_flights + 1
                    
                elif float (t[23]) == 0.00 or len(t[23]) == 0 :
                    #print('On Time', airport)
                    ontime_flights= (ontime_flights + 1)
                    
            except :
                #print('No Data', airport)
                nodata_flights = (nodata_flights + 1)

    return late_flights, ontime_flights, nodata_flights

#This function then takes the return from the previous function and creates a graph based on the data given

def graph_flights(late_flights,ontime_flights,nodata_flights):
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    data = ['Late', 'On time', 'No Data']
    flight_times = [late_flights, ontime_flights,nodata_flights]
    ax.bar(data,flight_times)
    ax.set_ylabel('Number of Flights')
    plt.title('2017-2018 Flight Timing')
    plt.show()

# This allows for the overall function to know what the airport variable means

airport = str(input('What airport'))
analyze_airports(airport)


```
